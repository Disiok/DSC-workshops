# 4. Joins and aggregation <a class="anchor" id="agg"></a>

### Exercise 4.1: Are there directors are also lead actors in movies. 
Right now we only care about the name and year of the movies they directed and acted.

````python
# Make a dataframe with ['movie_title', 'title_year'] indexed by 'director_name'
# .set_index()
# .ix[?, ?]

df_director = df_movies.set_index('director_name')
df_director = df_director.ix[:, ['movie_title', 'title_year']]
df_director = df_director.dropna()
df_director.head()
````

````python
# Make a dataframe with ['movie_title', 'title_year'] indexed by 'actor_1_name'
# .set_index()
# .ix[?, ?]

df_actor = df_movies.set_index('actor_1_name')
df_actor = df_actor.ix[:, ['movie_title', 'title_year']]
df_actor = df_actor.dropna()
df_actor.head()
````

````python
# Inner join the two dataframes we prepared
# .join()

df_joined = df_actor.join(df_director, lsuffix='_acted', rsuffix='_directed', how='inner')
df_joined.head()
````

### Exercise 4.2:  Just how many of these actor/director are there? Did they start out acting or directing?

````python
# len()
# .groupby()

len(df_joined.groupby(df_joined.index))
````

Let's look at one acting/directing movie pair for each person

````python
# .first()

df_first = df_joined.groupby(df_joined.index).first()
df_first.head()
````

````python
# .mean()
df_mean = df_joined.groupby(df_joined.index).mean()

df_mean.head()
````

People say directing is more prestigeous than acting. Do we see peole moving from acting to directing?

````python
# df[? > ?]
# .shape

df_a2d = df_mean[df_mean['title_year_directed'] > df_mean['title_year_acted']]
(df_a2d.shape[0], df_mean.shape[0])
````

# 6. Time Series <a class="anchor" id="time"></a>

````python
df_movies = pd.read_csv('movies.csv', parse_dates=['title_year'])
````

````python
df_year_gross = df_movies.ix[:, ['title_year', 'gross']]
df_year_gross = df_year_gross.dropna().set_index('title_year')
df_year_gross.head()
````

````python
df_10_year = df_year_gross.resample('10AS').mean()
df_10_year.plot(kind='bar', x=df_10_year.index.year)
````

````python
df_after_2000 = df_year_gross[df_year_gross.index.year >= 2000]

plt.xticks(rotation=90)
graph = sns.violinplot(data=df_after_2000, x=df_after_2000.index.year, y='gross')
````
